```python
import pandas as pd
import json
```

```python
# Load json file
file_path = "../data/json/smartsupply_data.json"

try:
    with open(file_path, 'r', encoding='utf-8') as file:
        data = json.load(file)
    print("Data type:", type(data))
    print(data.keys())
except FileNotFoundError:
    print(f"Error: The file '{file_path}' was not found.")
except json.JSONDecodeError:
    print(f"Error: Could not decode JSON from the file '{file_path}'. Check file format.")
```

```python
# Convert to pandas DataFrame
products_df = pd.DataFrame(data['products'])
products_df
```

```python
inventory_df = pd.DataFrame(data['inventory'])
inventory_df
```

```python
warehouses_df = pd.DataFrame(data['warehouses'])
warehouses_df
```

```python
movements_df = pd.DataFrame(data['movements'])
movements_df
```

```python
movements_df.columns
```

```python
# Check metadata
print("Metadata:", data.get('metadata', 'No metadata'))
```

```python
# Basic data analysis

# Products analysis
print("=== PRODUCTS ANALYSIS ===")
print(f"Total products: {len(products_df)}")
print(f"Categories: {products_df['category'].unique()}")
print(f"Price range: ${products_df['unit_price'].min()} - ${products_df['unit_price'].max()}")
print(f"Average price: ${products_df['unit_price'].mean():.2f}")
print()

# Warehouses analysis
print("=== WAREHOUSES ANALYSIS ===")
print(f"Total warehouses: {len(warehouses_df)}")
print(f"Regions: {warehouses_df['region'].unique()}")
print(f"Total capacity: {warehouses_df['capacity'].sum()} units")
print()

# Inventory analysis
print("=== INVENTORY ANALYSIS ===")
print(f"Total inventory records: {len(inventory_df)}")
print(f"Total quantity across all warehouses: {inventory_df['quantity'].sum()}")
print(f"Average quantity per product per warehouse: {inventory_df['quantity'].mean():.1f}")
low_stock = inventory_df[inventory_df['quantity'] <= inventory_df['reorder_level']]
print(f"Products at or below reorder level: {len(low_stock)}")
print()

# Movements analysis
print("=== MOVEMENTS ANALYSIS ===")
print(f"Total movements: {len(movements_df)}")
print(f"Movement types: {movements_df['movement_type'].unique()}")
print(f"Date range: {movements_df['timestamp'].min()} to {movements_df['timestamp'].max()}")
print(f"Total value of movements: ${movements_df['total_value'].sum():.2f}")
sales = movements_df[movements_df['movement_type'] == 'SALE']
print(f"Total sales value: ${sales['total_value'].sum():.2f}")
transfers = movements_df[movements_df['movement_type'] == 'TRANSFER']
print(f"Total transfer value: ${transfers['total_value'].sum():.2f}")
damages = movements_df[movements_df['movement_type'] == 'DAMAGE']
print(f"Total damage value: ${damages['total_value'].sum():.2f}")
```

```python
# Detailed analysis

# Top products by sales
print("=== TOP PRODUCTS BY SALES VALUE ===")
product_sales = sales.groupby('product_name')['total_value'].sum().sort_values(ascending=False)
for product, value in product_sales.head(5).items():
    print(f"{product}: ${value:.2f}")
print()

# Inventory by warehouse
print("=== INVENTORY BY WAREHOUSE ===")
warehouse_inventory = inventory_df.groupby('warehouse_name')['quantity'].sum().sort_values(ascending=False)
for warehouse, qty in warehouse_inventory.items():
    print(f"{warehouse}: {qty} units")
print()

# Movement types distribution
print("=== MOVEMENT TYPES DISTRIBUTION ===")
movement_counts = movements_df['movement_type'].value_counts()
for movement_type, count in movement_counts.items():
    print(f"{movement_type}: {count} movements")
print()

# Recent movements (last 7 days)
print("=== RECENT MOVEMENTS (LAST 7 DAYS) ===")
movements_df['timestamp'] = pd.to_datetime(movements_df['timestamp'])
recent_movements = movements_df[movements_df['timestamp'] >= pd.Timestamp.now() - pd.Timedelta(days=7)]
print(f"Recent movements: {len(recent_movements)}")
print(f"Recent sales value: ${recent_movements[recent_movements['movement_type'] == 'SALE']['total_value'].sum():.2f}")
```

## Attributes

```python
# extract the attributes
for table_name, table in data.items():
    if table_name == 'metadata':
        continue

    print("-" * 10)
    print(f"{table_name}")
    print("-" * 10)
    print("Attributes:")

    all_keys = set()
    for row in table:
        all_keys.update(row.keys())

    for key in all_keys:
        # gather non-null types for this key
        types = {type(row[key]) for row in table if key in row and row[key] is not None}

        # fallback if all values are None
        if not types:
            types = {type(None)}
        print(key, " | ", ", ".join(t.__name__ for t in types))
    print()
```
